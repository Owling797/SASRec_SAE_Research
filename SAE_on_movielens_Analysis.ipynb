````python
import re
import pandas as pd
import matplotlib.pyplot as plt

#change if needed
log_path = "/content/RecSAE_replica/log/SASRec_SAE/SASRec_SAE__MovieLens_1M/ML_1MTOPK__0__lr=0.0001__l2=1e-06__emb_size=64__num_layers=1__num_heads=1__sae_lr=0.0005__sae_k=32__sae_scale_size=32__batch_size=8.txt"

with open(log_path, "r") as f:
    log_lines = f.readlines()
````

````python
epoch_pattern = re.compile(r"Epoch\s+(\d+)")
loss_pattern = re.compile(r"loss=([\d.]+)")
dead_latent_pattern = re.compile(r"dead_latent=([\d.]+)")
metrics_pattern = re.compile(r'\(HR@5:([\d.]+),NDCG@5:([\d.]+),HR@10:([\d.]+),NDCG@10:([\d.]+),HR@20:([\d.]+),NDCG@20:([\d.]+),HR@50:([\d.]+),NDCG@50:([\d.]+)\)')
best_epoch_pattern = re.compile(r'Best Iter\(dev\)=\s*(\d+)')
````

````python
# Data storage
epochs, losses, dead_latents = [], [], []
hr5s, ndcg5s, hr10s, ndcg10s = [], [], [], []
hr20s, ndcg20s, hr50s, ndcg50s = [], [], [], []

best_epoch = None
recmodel_ndcg5s = None
````

````python
for line in log_lines:
    if 'Epoch' in line:
        epoch = int(epoch_pattern.search(line).group(1))
        loss = float(loss_pattern.search(line).group(1))
        dead_latent = float(dead_latent_pattern.search(line).group(1))

        epochs.append(epoch)
        losses.append(loss)
        dead_latents.append(dead_latent)

    elif line.strip().startswith("dev="):
        m = metrics_pattern.search(line)
        if m:
            hr5s.append(float(m.group(1)))
            ndcg5s.append(float(m.group(2)))
            hr10s.append(float(m.group(3)))
            ndcg10s.append(float(m.group(4)))
            hr20s.append(float(m.group(5)))
            ndcg20s.append(float(m.group(6)))
            hr50s.append(float(m.group(7)))
            ndcg50s.append(float(m.group(8)))

    elif "Best Iter(dev)=" in line:
        best_epoch = int(best_epoch_pattern.search(line).group(1))

    elif "[Rec] Dev Before Training:" in line:
        m = metrics_pattern.search(line)
        if m:
            recmodel_ndcg5s = float(m.group(2))

    elif "Test After Training:" in line:
        epochs.append("RecSAE_test")
        losses.append(0)
        dead_latents.append(0)
        m = metrics_pattern.search(line)
        if m:
            hr5s.append(float(m.group(1)))
            ndcg5s.append(float(m.group(2)))
            hr10s.append(float(m.group(3)))
            ndcg10s.append(float(m.group(4)))
            hr20s.append(float(m.group(5)))
            ndcg20s.append(float(m.group(6)))
            hr50s.append(float(m.group(7)))
            ndcg50s.append(float(m.group(8)))

    elif "[Rec] Test Before Training:" in line:
        epochs.append("SASRec_test")
        losses.append(0)
        dead_latents.append(0)
        m = metrics_pattern.search(line)
        if m:
            hr5s.append(float(m.group(1)))
            ndcg5s.append(float(m.group(2)))
            hr10s.append(float(m.group(3)))
            ndcg10s.append(float(m.group(4)))
            hr20s.append(float(m.group(5)))
            ndcg20s.append(float(m.group(6)))
            hr50s.append(float(m.group(7)))
            ndcg50s.append(float(m.group(8)))
````

````python
df = pd.DataFrame({
    "Epoch": epochs,
    "Loss": losses,
    "Dead Latent": dead_latents,
    "HR@5": hr5s,
    "NDCG@5": ndcg5s,
    "HR@10": hr10s,
    "NDCG@10": ndcg10s,
    "HR@20": hr20s,
    "NDCG@20": ndcg20s,
    "HR@50": hr50s,
    "NDCG@50": ndcg50s
})

df
````

````python
numeric_df = df[df["Epoch"].apply(lambda x: isinstance(x, int))]

fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.plot(numeric_df["Epoch"], numeric_df["Loss"], color="blue", marker="o", label="Train Loss")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss", color="blue")
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(0, 0.05)

if best_epoch is not None:
    best_loss = numeric_df[numeric_df["Epoch"] == best_epoch]["Loss"].values[0]
    ax1.annotate(f"Best Epoch {best_epoch}", xy=(best_epoch, best_loss),
                 xytext=(best_epoch, best_loss + 0.01),
                 arrowprops=dict(facecolor='blue', arrowstyle='->'),
                 color='blue')

ax2 = ax1.twinx()
ax2.plot(numeric_df["Epoch"], numeric_df["NDCG@5"], color="orange", marker="o", label="NDCG@5")
ax2.set_ylabel("NDCG@5", color="orange")
ax2.tick_params(axis='y', labelcolor='orange')
ax2.set_ylim(0.045, 0.075)

if recmodel_ndcg5s:
    ax2.axhline(recmodel_ndcg5s, color='orange', linestyle='--', label='SASRec-Dev NDCG@5')

plt.title("RecSAE Training Curve")
plt.show()
````

````python
import ast
import re

def parse_array(s):
    # Remove brackets and split by whitespace
    return np.array(re.findall(r'\d+\.\d+|\d+', s), dtype=float if '.' in s else int)

df['history'] = df['history'].apply(parse_array)
df['indices'] = df['indices'].apply(parse_array)
df['values'] = df['values'].apply(parse_array)

df.head()
````

## Find activation from movie

````python
from collections import defaultdict, Counter
import pandas as pd

def movie_to_concept_and_related_movies(movie_id, df, concept_to_items, meta_df, top_n=10):
    # Step 1: Find concepts that activate this movie and how often
    concept_counts = Counter()

    for _, row in df.iterrows():
        if movie_id in row['history']:
            for idx in row['indices']:
                concept_counts[idx] += 1

    if not concept_counts:
        print(f"No activations found for movie ID {movie_id}")
        return pd.DataFrame()

    # Step 2: Get the most activating concept
    top_concept = concept_counts.most_common(1)[0][0]
    print(f" Top concept that activated movie {movie_id}: {top_concept}")

    # Step 3: Get all movies this concept activated
    all_items = concept_to_items[top_concept]
    item_counts = Counter(all_items)

    # Step 4: Convert to DataFrame
    related_df = meta_df[meta_df['item_id'].isin(item_counts)].copy()
    related_df['activation_count'] = related_df['item_id'].apply(lambda x: item_counts[x])
    related_df = related_df.sort_values(by='activation_count', ascending=False)

    # Step 5: Optional  mark input movie
    related_df['is_target_movie'] = related_df['item_id'] == movie_id

    return related_df[['item_id', 'title', 'genre_list', 'activation_count', 'is_target_movie']].head(top_n)
````
